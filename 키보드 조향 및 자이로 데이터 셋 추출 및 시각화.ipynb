```python
import pygame
import time
from pop import Pilot

bot = Pilot.SerBot()

# reset
pygame.init()

# window settings
screen_width = 640
screen_height = 480
screen = pygame.display.set_mode((screen_width, screen_height))

#자이로 값 저장할 리스트
stop_value = []
forward_value = []
backward_value = []
left_value = []
right_value = []

# 키보드를 이용한 방향 코드
while True:
    # check for key events
    value = bot.getGyro()
    #print(value)
    #time.sleep(1)
    keys = pygame.key.get_pressed()
    if keys[pygame.K_UP]:
        # move forward
        #bot.move(0, 30)
        if keys[pygame.K_RIGHT]:
            bot.move(45, 30)
        elif keys[pygame.K_LEFT]:
            bot.move(315, 30)
        else:
            bot.move(0, 30)
            value = bot.getGyro()
            forward_value.append(value)
    elif keys[pygame.K_DOWN]:
        # move backward
        #bot.move(180, 30)
        if keys[pygame.K_RIGHT]:
            bot.move(135, 30)
        elif keys[pygame.K_LEFT]:
            bot.move(215, 30)
        else:
            bot.move(180, 30)
            value = bot.getGyro()
            backward_value.append(value)
    elif keys[pygame.K_LEFT]:
        # turn left
        bot.move(270, 30)
        value = bot.getGyro()
        left_value.append(value)
    elif keys[pygame.K_RIGHT]:
        # turn right
        bot.move(90, 30)
        value = bot.getGyro()
        right_value.append(value)
    elif keys[pygame.K_q]: #조향 이동
        bot.forward()
        bot.steering = -1.0
    elif keys[pygame.K_w]:
        bot.forward()
        bot.steering = 1.0
    elif keys[pygame.K_a]:
        bot.backward()
        bot.steering = -1.0
    elif keys[pygame.K_s]:
        bot.backward()
        bot.steering = 1.0
        
    elif keys[pygame.K_SPACE]: #자이로 값 출력
        print('stop gyro : ', stop_value)
        print('forward gyro : ', forward_value)
        print('backward gyro : ', backward_value)
        print('left gyro : ', left_value)
        print('right gyro : ', right_value)
        pygame.quit()
        #quit()
    
    else:
        # 멈춤
        bot.stop()
        value = bot.getGyro()
        stop_value.append(value)
        
        # pygame 종료 후 각 방향으로 저장된 자이로 값 출력 
    for event in pygame.event.get():
        if event.type == pygame.K_ESCAPE:
            print('stop gyro : ', stop_value)
            print('forward gyro : ', forward_value)
            print('backward gyro : ', backward_value)
            print('left gyro : ', left_value)
            print('right gyro : ', right_value)
            pygame.quit()
            quit()
```
